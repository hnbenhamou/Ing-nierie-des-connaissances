## **2-Importation des bibliothèques**

```python
import spacy
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations
```

## **3-4-Chargement du modele de langage et du texte**


*   Extraction des entités et relation des phrases



```python
text = """
Google is a global technology company that specializes in internet-related services and products.
NASA, the National Aeronautics and Space Administration, is the United States government agency responsible for the civilian space program.
Elon Musk is the CEO of SpaceX and Tesla, and he is known for his ventures in technology and space exploration.
Google and NASA have collaborated on various projects, including advancements in artificial intelligence and space exploration.
Elon Musk's company SpaceX has a partnership with NASA for launching rockets to the International Space Station.
"""
doc = nlp(text)
```

```python
entities = []
relations = []
for sentence in doc.sents:
    ents_in_sentence = [ent.text for ent in sentence.ents]
    entities.extend(ents_in_sentence)
    for ent1, ent2 in combinations(ents_in_sentence, 2):
        relations.append((ent1, ent2))
entities = list(set(entities))
relations = list(set(relations))
```

## **5-Création du graphe sémantique**

```python
G = nx.Graph()
G.add_nodes_from(entities)
G.add_edges_from(relations)
```

## **6-Visualition et analyse du réseau sémantique**

```python
import matplotlib.pyplot as plt
import networkx as nx

# Vérifiez que le graphe 'G' contient bien des nœuds et des arêtes
print("Nombre de nœuds :", G.number_of_nodes())
print("Nombre d'arêtes :", G.number_of_edges())

# Créer la figure et afficher le graphe
plt.figure(figsize=(10, 8))
nx.draw(
    G,
    with_labels=True,
    node_size=1500,
    node_color="lightblue",
    font_size=10,
    font_weight="bold"
)
plt.title("Réseau Sémantique")
plt.show()

```

```python
import spacy
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Charger le modèle de spaCy
nlp = spacy.load("en_core_web_sm")

# Charger le texte à partir d'un fichier
with open("AI.txt", "r") as file:
    text = file.read()

# Analyser le texte avec spaCy
doc = nlp(text)

# Extraire les entités et relations par co-occurrence dans la même phrase
entities = []
relations = []
for sentence in doc.sents:
    ents_in_sentence = [ent.text for ent in sentence.ents]
    entities.extend(ents_in_sentence)
    for ent1, ent2 in combinations(ents_in_sentence, 2):
        relations.append((ent1, ent2))

# Éliminer les doublons
entities = list(set(entities))
relations = list(set(relations))

# Créer le graphe non orienté
G = nx.Graph()
G.add_nodes_from(entities)
G.add_edges_from(relations)

# Visualiser le graphe
plt.figure(figsize=(12, 12))
nx.draw_networkx(G, with_labels=True, node_size=700, font_size=10, node_color='skyblue')
plt.show()
```

## **2-** Explorer des méthodes avancées de détection de relations, telles que les dépendances syntaxiques

```python
entities = []
relations = []

for sentence in doc.sents:
    ents_in_sentence = [ent.text for ent in sentence.ents]
    entities.extend(ents_in_sentence)

    # Parcourir chaque token de la phrase pour détecter les relations SVO
    for token in sentence:
        if token.dep_ == "nsubj" and token.head.pos_ == "VERB":
            subject = token.text
            verb = token.head.text
            for child in token.head.children:
                if child.dep_ == "dobj":
                    obj = child.text
                    relations.append((subject, obj))

# Éliminer les doublons
entities = list(set(entities))
relations = list(set(relations))

# Création et visualisation du graphe avec les relations syntaxiques
G = nx.Graph()
G.add_nodes_from(entities)
G.add_edges_from(relations)

plt.figure(figsize=(12, 12))
nx.draw_networkx(G, with_labels=True, node_size=700, font_size=10, node_color='lightgreen')
plt.show()
```

## **3-** Appliquer une mesure de similarité pour connecter les concepts liés sémantiquementmais non explicitement co-mentionnés.

```python
# Calculer les similarités entre entités et ajouter une relation si elles dépassent un seuil
similarity_threshold = 0.75  # Définir un seuil de similarité

similar_relations = []
entity_vectors = [(ent, nlp(ent)) for ent in entities]

for (ent1, vec1), (ent2, vec2) in combinations(entity_vectors, 2):
    if vec1.similarity(vec2) > similarity_threshold:
        similar_relations.append((ent1, ent2))

# Ajouter ces relations similaires au graphe
G.add_edges_from(similar_relations)

# Visualiser le graphe avec des connexions similaires ajoutées
plt.figure(figsize=(12, 12))
nx.draw_networkx(G, with_labels=True, node_size=700, font_size=10, node_color='lightcoral')
plt.show()
```

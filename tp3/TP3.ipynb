```python
# Importation des bibliothèques nécessaires
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Création du graphe
G = nx.DiGraph()
```

```python
# Ajout des nœuds et des relations entre eux
G.add_edge("Chien", "Animal", relation="est un")
G.add_edge("Chien", "Viande", relation="mange")
G.add_edge("Viande", "Nourriture", relation="est un")
G.add_edge("Chat", "Animal", relation="est un")
G.add_edge("Animal", "Être vivant", relation="est un")
G.add_edge("Animal", "Respirer", relation="a besoin de")
```

```python
# Dessin du graphe
pos = nx.spring_layout(G)  # Positionnement des nœuds pour la visualisation
plt.figure(figsize=(10, 8))
```

```python
# Dessiner les nœuds
nx.draw_networkx_nodes(G, pos, node_color="lightblue", node_size=1500)

# Dessiner les arêtes avec des labels pour les relations
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=20, edge_color="gray")
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")

# Ajouter les labels pour les relations (arêtes)
edge_labels = {(u, v): d["relation"] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red")


# Afficher le graphe
plt.title("Réseau Sémantique de Concepts et Relations")
plt.axis("off")  # Désactiver les axes
plt.show()
```

```python
# Recherche de connexions pour le concept "Chien"
print("\nRelations associées au concept 'Chien' :")
for neighbor in G.neighbors("Chien"):
    print(f"Chien -> {neighbor}: {G['Chien'][neighbor]['relation']}")
```

```python
# Afficher la relation entre "Chien" et "Animal" s'il existe
if G.has_edge("Chien", "Animal"):
    print(f"\nRelation entre 'Chien' et 'Animal' : {G['Chien']['Animal']['relation']}")
else:
    print("\nAucune relation directe entre 'Chien' et 'Animal'.")
```
